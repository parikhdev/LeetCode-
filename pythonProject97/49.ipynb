```python
cs_df=pd.read_csv("computersales.csv")
cs_df
```

```python
# mean profit of computer sales
cs_df["Profit"].mean()
# product with highest profit
cs_df.loc[cs_df["Profit"].idxmax(),["Product ID","Contact"]]
# no of people who purchased from WV
cs_df[cs_df["State"]=="WV"]["Sale ID"].count()
# no of purchases in 2019
cs_df[cs_df["Year"]==2019]["Sale ID"].count()
# no of sales for each product type
cs_df["Product ID"].value_counts()
# list of contact of customers who bought specific prd
cs_df[cs_df["Product ID"]=="MY2J2LL"][["Contact","State"]]
# how many people made a website purchase with profit over $150
cs_df[(cs_df["Lead"]=="Website") & (cs_df["Profit"]>150)]["Sale ID"].count()
```
