```python
temp=arr[:,2]
mean=np.mean(temp)
std=np.std(temp)
z = (temp - mean) / std
anomalies = temp[np.abs(z) > 2]
anomalies
```

```python
import numpy as np

temps = np.array([25, 27, 29, 30, 32, 35])

changes = np.diff(temps)

print("Daily Changes:", changes)
print("Average Increase:", np.mean(changes))
```

```python
import numpy as np

city_a = np.array([30, 32, 35, 33, 31])
city_b = np.array([-9, 31, 5, 43, -3])

corr = np.corrcoef(city_a, city_b)[0,1]

print("Similarity:", corr)
```
