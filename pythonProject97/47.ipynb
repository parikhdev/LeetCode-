```python
import numpy as np
temp=np.array([34,39,18,34,31,100,34,30,37])
mean=np.mean(temp)
std=np.std(temp)
print(mean)
```

```python
print(std)
```

```python
nor=(temp-mean)/std
nor
```

```python
temp[np.abs(nor)>2]
```
