```python
# Lets print table of any number
 
def table(n):
    print(f"The Table of {n} is: ")
    for i in range(1,11):
        print(f"{n} x {i} = {n*i}")
    print()
numbers = input("Enter the numbers you want to the print the table separated by spaces: ").split(" ")
nums = list(map(int, numbers))

for num in nums:
    table(num)
```
