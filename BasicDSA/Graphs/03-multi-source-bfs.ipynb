## Multi-Source Breadth First Search

### Graph Creation

```python
graph = {
    10: [40, 20],
    40: [10, 35],
    35: [40, 20, 50, 25],
    50: [35],
    25: [35],
    20: [10, 35, 60, 80, 70],
    60: [20, 80, 90, 70],
    80: [20, 60, 70],
    70: [20, 80, 60],
    90: [60]
}
```

### Traversal

```python
import math
from collections import deque
```

```python
def multi_source_breadth_first_search(graph, sources):
    distance = {node: math.inf for node in graph}
    queue = deque()

    for source in sources:
        distance[source] = 0
        queue.append(source)
    
    while queue:
        curr = queue.popleft()
        for neighbour in graph[curr]:
            if distance[neighbour] == math.inf:
                distance[neighbour] = distance[curr] + 1
                queue.append(neighbour)
    
    return distance
```

```python
multi_source_breadth_first_search(graph, [10, 35, 70])
```
