## Longest Path DAG

### Setup

```python
from collections import defaultdict
```

```python
graph = defaultdict(list)

graph[6] = [1, 5]
graph[0] = [1, 2]
graph[1] = [2, 5]
graph[2] = [3]
graph[5] = [3, 4]
graph[3] = []
graph[4] = []

weights = {
    (6, 1): 2,
    (6, 5): 4,
    (0, 1): 1,
    (0, 2): 5,
    (1, 2): 1,
    (1, 5): 3,
    (2, 3): 2,
    (5, 3): -2,
    (5, 4): 1
}

src = 6

for u in graph:
    print(u, "->", graph[u])
```

### Algorithm

```python
def topo_dfs(graph):
    visited = set()
    topo = []

    def dfs(u):
        visited.add(u)
        for v in graph[u]:
            if v not in visited:
                dfs(v)
        topo.append(u)
    
    for node in graph:
        if node not in visited:
            dfs(node)
    
    return topo[::-1]
```

```python
def longest_path(graph):
    topo = topo_dfs(graph)
    res = {node: 0 for node in graph}

    for u in topo:
        for v in graph[u]:
            res[v] = max(res[v], res[u] + 1)

    return res
```
