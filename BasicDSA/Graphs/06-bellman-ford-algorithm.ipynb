## Bellman-Ford Algorithm

### Graph Creation

```python
from collections import defaultdict
```

```python
V = 7

# Edge list: (u, v, weight) / (u, v, w)
edges = [
    (1, 2, 6),
    (1, 3, 5),
    (1, 4, 5),

    (4, 3, -2),
    (3, 2, -2),

    (2, 5, -1),
    (3, 5, 1),

    (5, 7, 3),
    (4, 6, -1),
    (6, 7, 3)
]

graph = defaultdict(list)

for u, v, w in edges:
    graph[u].append((v, w))

# Print adjacency list
for node in sorted(graph):
    print(f"{node} -> {graph[node]}")
```

### Traversal

```python
import math
```

```python
def bellman_ford(V, edges, source):
    distances = [math.inf] * (V + 1)
    distances[source] = 0

    for i in range(V - 1):
        print(f'Iteration {i + 1}:')
        for u, v, w in edges:
            if (distances[u] != math.inf) and (distances[u] + w < distances[v]):
                distances[v] = distances[u] + w
                print(f'Updated distances[{v}] to {distances[v]}')
        print('Distances:', distances[1:])
        print('-' * 40)
    
    for u, v, w in edges:
        if (distances[u] != math.inf) and (distances[u] + w < distances[v]):
            print('Negative Cycle Detected!!')
            return None

    return distances[1:]
```

```python
bellman_ford(V, edges, 1)
```
