## Floyd-Warshall Algorithm

### Graph Creation

```python
INF = float('inf')

# Number of nodes
n = 4

# Initialize adjacency matrix
matrix = [[INF] * n for _ in range(n)]

# Distance from node to itself = 0
for i in range(n):
    matrix[i][i] = 0

# Add edges
edges = [
    (1, 2, 4),
    (2, 1, 10),
    (2, 3, 3),
    (3, 4, 2),
    (4, 1, 1),
    (1, 4, 9),
    (3, 1, 5)
]

for u, v, w in edges:
    matrix[u-1][v-1] = w

for row in matrix:
    print(row)
```

### Traversal

```python
def floyd_warshall(matrix):
    n = len(matrix)

    for k in range(n):
        for i in range(n):
            for j in range(n):
                if matrix[i][k] + matrix[k][j] < matrix[i][j]:
                    matrix[i][j] = matrix[i][k] + matrix[k][j]
    
    return matrix
```

```python
res = floyd_warshall(matrix)

for row in res:
    print(row)
```
