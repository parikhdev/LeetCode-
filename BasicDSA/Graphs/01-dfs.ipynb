## Depth First Search

### Graph Creation

```python
graph = {
    10: [40, 20],
    40: [10, 35],
    35: [40, 20, 50, 25],
    50: [35],
    25: [35],
    20: [10, 35, 60, 80, 70],
    60: [20, 80, 90],
    80: [20, 60, 70],
    70: [20, 80],
    90: [60]
}
```

### Traversal

```python
def depth_first_search(node, visited):
    visited.add(node)
    print(node, end=' ')

    for neighbour in graph[node]:
        if neighbour not in visited:
            depth_first_search(neighbour, visited)
```

```python
visited = set()
start_node = 10

print(f'DFS traversal starting at node {start_node}:')
depth_first_search(start_node, visited)
```
