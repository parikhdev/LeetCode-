## Breadth First Search

### Graph Creation

```python
graph = {
    10: [40, 20],
    40: [10, 35],
    35: [40, 20, 50, 25],
    50: [35],
    25: [35],
    20: [10, 35, 60, 80, 70],
    60: [20, 80, 90],
    80: [20, 60, 70],
    70: [20, 80],
    90: [60]
}
```

### Traversal

```python
from collections import deque
```

```python
def breadth_first_search(node):
    visited = set()
    visited.add(node)
    queue = deque([node])

    while queue:
        curr = queue.popleft()
        print(curr, end=' ')

        for neighbour in graph[curr]:
            if neighbour not in visited:
                visited.add(neighbour)
                queue.append(neighbour)
```

```python
breadth_first_search(10)
```
