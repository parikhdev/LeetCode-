## Dijkstra's Algorithm

### Graph Creation

```python
graph = {
    10: [(40,1), (20,1)],
    40: [(10,1), (35,1)],
    35: [(40,1), (50,1), (25,1), (20,1)],
    50: [(35,1)],
    25: [(35,1)],
    
    20: [(10,1), (35,1), (60,0), (80,1), (70,1)],
    60: [(20,0), (90,1), (80,0), (70,0)],
    90: [(60,1)],
    80: [(60,0), (20,1), (70,1)],
    70: [(60,0), (20,1), (80,1)]
}
```

### Traversal

```python
import math
import heapq
```

```python
def dijkstras(graph, source):
    distances ={node: math.inf for node in graph}
    distances[source] = 0

    pq = [(0, source)]

    while pq:
        curr_dist, curr_node = heapq.heappop(pq)

        if curr_dist > distances[curr_node]:
            continue
        
        for neighbour, weight in graph[curr_node]:
            new_distance = distances[curr_node] + weight
            if new_distance < distances[neighbour]:
                distances[neighbour] = new_distance
                heapq.heappush(pq, (new_distance, neighbour))
    
    return distances
```

```python
dijkstras(graph, 10)
```
