## Traversals on Trees

### 0. Creating the Tree

```python
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None
```

```python
# nodes
A = TreeNode('A')
B = TreeNode('B')
C = TreeNode('C')
D = TreeNode('D')
E = TreeNode('E')
F = TreeNode('F')

# edges
A.left = B
A.right = C

B.left = D
B.right = E

C.left = F
```

### 1. Preorder Traversal

```python
def preorder_traversal(node):
    if not node:
        return
    
    print(node.val, end=' ')
    preorder_traversal(node.left)
    preorder_traversal(node.right)
```

```python
preorder_traversal(A)
```

### 2. Inorder Traversal

```python
def inorder_traversal(node):
    if not node:
        return
    
    inorder_traversal(node.left)
    print(node.val, end=' ')
    inorder_traversal(node.right)
```

```python
inorder_traversal(A)
```

### 3. Postorder Traversal

```python
def postorder_traversal(node):
    if not node:
        return
    
    postorder_traversal(node.left)
    postorder_traversal(node.right)
    print(node.val, end=' ')
```

```python
postorder_traversal(A)
```

### 4. Level Order Traversal

```python
from collections import deque
```

```python
def level_order_traversa(node):
    if not node:
        return
    
    q = deque([node])
    while q:
        curr = q.popleft()
        print(curr.val, end=' ')

        if curr.left:
            q.append(curr.left)
        if curr.right:
            q.append(curr.right)
```

```python
level_order_traversa(A)
```
