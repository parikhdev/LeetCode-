## Coding Problems on Trees

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

### 1. Maximum Depth of a Binary Tree

- Given the root of a binary tree, return its maximum depth

```python
root = TreeNode(40)
root.left = TreeNode(30)
root.right = TreeNode(50)
root.left.left = TreeNode(25)
root.left.right = TreeNode(35)
root.right.left = TreeNode(45)
root.right.right = TreeNode(60)
```

```python
def max_depth(node):
    if not node:
        return 0
    
    left = max_depth(node.left)
    right = max_depth(node.right)
    return 1 + max(left, right)
```

```python
max_depth(root) - 1
```

### 2. Count Number of Nodes in a Binary Tree

```python
def count_nodes(node):
    if not node:
        return 0
    return 1 + count_nodes(node.left) + count_nodes(node.right)
```

```python
count_nodes(root)
```

### 3. Sum of All Nodes in Binary Tree

```python
def tree_sum(node):
    if not node:
        return 0
    
    left_sum = tree_sum(node.left)
    right_sum = tree_sum(node.right)
    return node.val + left_sum + right_sum
```

```python
tree_sum(root)
```

### 4. Validate a Binary Search Tree

```python
import math
```

```python
def is_valid_bst(root):
    def traverse(node, low, high):
        if not node:
            return True

        if not (low < node.val < high):
            return False

        return (
            traverse(node.left, low, node.val) and traverse(node.right, node.val, high)
        )
    return traverse(root, -math.inf, math.inf)    
```

```python
is_valid_bst(root)
```

### 5. Lowest Common Ancestor in a BST

- Given a BST and two nodes `p` and `q`, find their lowest common ancestor

```python
# Create nodes
root = TreeNode(6)

root.left = TreeNode(2)
root.right = TreeNode(8)

root.left.left = TreeNode(0)
root.left.right = TreeNode(4)

root.right.left = TreeNode(7)
root.right.right = TreeNode(9)

root.left.right.left = TreeNode(3)
root.left.right.right = TreeNode(5)
```

```python
def lowest_common_ancestor(root, p, q):
    while root:
        if p.val < root.val and q.val < root.val:
            root = root.left
        elif p.val > root.val and q.val > root.val:
            root = root.right
        else:
            return root
```

```python
lowest_common_ancestor(root, p=root.left, q=root.right).val
```

```python
lowest_common_ancestor(root, p=root.left, q=root.left.right).val
```

```python
lowest_common_ancestor(
    root,
    p=root.left.right.left,
    q=root.right.left
).val
```

### 6. Kth Smallest Element in a BST

```python
root = TreeNode(40)
root.left = TreeNode(30)
root.right = TreeNode(50)
root.left.left = TreeNode(25)
root.left.right = TreeNode(35)
root.right.left = TreeNode(45)
root.right.right = TreeNode(60)
```

```python
def kth_smallest(node, k):
    stack = []

    while True:
        while node:
            stack.append(node)
            node = node.left

        node = stack.pop()
        k -= 1

        if k == 0:
            return node.val
        
        node = node.right
```

```python
kth_smallest(root, 3)
```

```python
kth_smallest(root, 5)
```

### 7. Find Maximum Value in an N-ary Tree

```python
class NAryNode:
    def __init__(self, val):
        self.val = val
        self.children = []
```

```python
root = NAryNode(10)

# level 1
node_20 = NAryNode(20)
node_30 = NAryNode(30)
node_40 = NAryNode(40)
root.children = [node_20, node_30, node_40]

# level 2
node_20.children = [NAryNode(50), NAryNode(60), NAryNode(70)]
node_30.children = [NAryNode(80)]
node_40.children = [NAryNode(90), NAryNode(100)]
```

```python
def find_max(node):
    if not node:
        return -math.inf

    res = node.val
    for child in node.children:
        res = max(res, find_max(child))
    
    return res
```

```python
find_max(root)
```

### 8. Height of N-ary Tree

```python
def height(node):
    if not node:
        return 0
    
    heights = []
    for child in node.children:
        heights.append(height(child))
    
    return 1 + max(heights, default=0)
```

```python
height(root)
```

### 9. Maximum Path Sum in a Binary Tree

- Find the maximum path sum in a binary tree
- A path can start and end at any node, but must go downwards through parent–child connections

```python
root = TreeNode(-10)
root.left = TreeNode(9)
root.right = TreeNode(20)

root.right.left = TreeNode(15)
root.right.right = TreeNode(7)
```

```python
def max_path_sum(root):
    max_sum = -math.inf

    def helper(node):
        nonlocal max_sum

        if not node:
            return 0
        
        left_gain = max(helper(node.left), 0)
        right_gain = max(helper(node.right), 0)

        current_path = node.val + left_gain + right_gain
        max_sum = max(max_sum, current_path)

        return node.val + max(left_gain, right_gain)

    helper(root)
    return max_sum
```

```python
max_path_sum(root)
```

### 10. Sum of Left Leaves

- Return the sum of all left leaf nodes in a binary tree

```python
root = TreeNode(3)
root.left = TreeNode(9)
root.right = TreeNode(20)

root.right.left = TreeNode(15)
root.right.right = TreeNode(7)
```

```python
def sum_left_leaves(root):
    if not root:
        return 0
    
    total = 0
    if root.left and not root.left.left and not root.left.right:
        total += root.left.val
    
    total += sum_left_leaves(root.left)
    total += sum_left_leaves(root.right)
    
    return total
```

```python
sum_left_leaves(root)
```
