```python
# Q1 Exam sets maker

import random
var = int(input("Enter number of sets you want: "))
num = int(input("Enter how many Questions you want in the set "))
Questions = ("5+6=", "4+9=","8+6=","5+0=","5+2=","5-6=","5*6=","5**6=","56-2=","59-9=","5%6=","6+0=","6+9=","5+1=","5-6=","5+6=","4+9=")

max_unique = len(set(Questions))
if num > max_unique:
    print(f"Warning: Only {max_unique} unique questions available. So only {max_unique} questions can be present in the set")
    num = max_unique

while var > 0:
    QSet = set()
    while len(QSet) < num:
        QSet.add(random.choice(Questions))
    
    final_List = list(QSet)
    random.shuffle(final_List)
    print(f"Set {var} is {final_List}")
    var -= 1
    


```

```python
# Q2 Classify names based on lengths 
names = input("Enter the names separated by spaces: ").split()
nameDict = {}
for name in names:
    n = len(name)
    if n not in nameDict.keys():
        nameDict[n] = [name]
    else:
        nameDict[n] = nameDict[n] + [name]
    print(nameDict)



```

```python
# Q3 Creating Registration, Login, Change Password using a dictionary

print("Welcome to our System.")
Users = {} 

while True:
    print("\n--- Main Menu ---")
    ch = input("Enter 'R' for Registration, 'L' for Login, 'A' for Admin, or 'E' for Exit: ").upper()
    
    if ch == 'R':
        print("\n[Registration Page]")
        username = input("Enter Your UserName: ")
        if username in Users:
            print("Username already exists. Please try with another name.")
        else:
            userpwd = input("Enter Your Password: ")
            Users[username] = userpwd
            print("Registration Successful!")

    elif ch == 'L':
        print("\n[Login Page]")
        name = input("Enter Your UserName: ")
        
        if name not in Users:
            print("Account does not exist. Please go to the Registration Page.")
        else:
            pwd = input("Enter your Password: ")
            if pwd != Users[name]:
                print("Wrong Password Entered.")
            else:
                print("You are logged in Successfully!")
                
                func1 = input("Do you want to change Password? Enter 'Y' for Yes or 'N' for No: ").upper()
                if func1 == 'Y':
                    NewPwd = input("Please Enter your NEW Password: ")
                    Users[name] = NewPwd
                    print("Your Password changed successfully!")
                elif func1 == 'N':
                    print("Password remains unchanged.")
                else:
                    print("Please type a Valid Character.")
        
    elif ch == 'A':
        print("\n[Admin View]")
        func2 = input("Do you want to see the DataBase? Enter 'Y' for Yes or 'N' for No: ").upper()
        if func2 == 'Y':
            SecretCode = "Secret123"
            AdminCode = input("Please Enter the Admin Code: ")
            if SecretCode == AdminCode:
                print(f"This is the DataBase: {Users}")
            else:
                print("Your Entered Admin Code is Wrong...")
        elif func2 == 'N':
            print("Admin view cancelled.")
        else:
            print("Please type a Valid Character.")
            
    elif ch == 'E':
        print("You have exited from the system. Goodbye!")
        break 
        
    else:
        print("Please type a Valid Character...")
```
