```python
#Q1 Print Perfect Squares from a list of numbers
import math
userList = input("Enter the numbers separated by spaces: ")
lst = list(map(int,userList.split()))

newlst = [i for i in lst if i>0 and math.isqrt(i)**2 == i]
# newlst = []
# for i in lst:
#    if i > 0:
#       root = math.isqrt(i)
#       if root*root == i:
#          newlst.append(i)
print(f"The Perfect Squares are {newlst}")
         
```

```python
# Q2 Print total bill of Customer, Prices = [] and Quantity = []
priceList = input("Enter Prices separated by coma: ")
QtList = input("Enter the Quantity of the items separated by coma: ")
price = list(map(int, priceList.split()))
Quant = list(map(int,QtList.split()))

TotalBill = sum(p * q for p, q in zip(price,Quant))
# TotalBill = sum(price[i] * Quant[i] for i in range(len(price)))
print(TotalBill)
```

```python
# Q3 Given a list = [roll no, name, m1,m2,m3,m4,m5] and print list = [roll no, name, total marks]
roll = input("Enter your Roll Number: ")
name = input("Enter your Name: ")
marks = input("Enter marks of five subjects with spaces: ")
lst = [roll, name, list(map(int, marks.split()))]

newlst = []
for i in lst:
    if isinstance(i,str):
        newlst.append(i)
    elif isinstance(i,list):
        newlst.append(sum(i))
print(newlst)
```

```python
# Q4 Remove Duplicates from a list
userlist = input("Enter the numbers separated by spaces: ")
num = list(map(int, userlist.split()))

newlst = []
for i in num:
    if i not in newlst:
        newlst.append(i)

print(f"list without duplicates are {newlst}")
```

```python
# Q5 Print largest word from the string based on length
sentence = input("Enter a sentence separated by spaces: ")
lst = sentence.split()

# largest = len(max(lst, key = len))
# words = [x for x in lst if len(x) == largest]

largest = 0
words = []
for i in lst:
    if largest < len(i):
        largest = len(i)
        words = [i]
    elif largest == len(i):
        words.append(i)

print(f"The largest word from the sentence '{sentence}' is '{words}")
```
