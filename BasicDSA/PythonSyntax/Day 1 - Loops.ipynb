```python
# Q1 Multiplicative Table
num = int(input("Enter any number: "))
print(f"This is the Table of {num}")
for i in range(1,11):
    print(f"{num} * {i} = {num * i}")
```

```python
# Q2 Sum of N numbers
num = int(input("Enter a number"))
sum = 0
print(f"The sum of {num} natural numbers are: ")
while num > 0:
    sum = sum + num
    num -= 1
print(sum)
```

```python
# Q3 List Filtering [12, 15, 32, 42, 55, 75, 122, 132, 150, 180, 200] If you hit a number greater than 150, stop the loop entirely.
nums = input("Enter a list of numbers separated by spaces ")
nums_list = list(map(int,nums.split()))
new_list = []
for i in nums_list:
    if i <= 150:
        if i%5==0:
            new_list.append(i)
    else:
        break
print(f" Given list is {nums_list} and The filtered list is {new_list}")
```

```python
# Q4 Digit Counter
num = int(input("Enter a number"))
count = 0
while num > 0:
    count += 1
    num = num // 10
print(f"The total digits are {count}")
```

```python
# Q5 Fibonacci Sequence
num = int(input("Enter a number of how many fibonacci sequence digits you want "))
Fib = [0,1]
new_dig = 0
if num > 0:
    for i in range(num):
        new_dig = Fib[i] + Fib[i+1]
        Fib.append(new_dig)
    print(f"The {num} digits of Fibonacci are {Fib[0:num]}")
else:
    print(f"Please Enter a number greater than 0")
```

```python
# Q6 Check if a number is Prime or not
num = int(input("Enter a number "))
digit = 2
if num > 2:
    while digit < num:
        if num % digit == 0:
            print(f"The given number {num} is not prime")
            break
        digit += 1
    else:
        print(f"Yes, the given number {num} is prime number")
elif num == 2:
    print(f"Yes, the given number {num} is prime")
else:
    print(f"The given number {num} is not prime")

```

```python
# Q7 Reverse the Integer
num = int(input("Enter a number "))
n = num
st = ""
dig = 0
while num > 0:
    dig = str(num%10)
    st += dig
    num = num // 10
print(f"The reverse of the number {n} is {st}")
```

```python
# Q8 The Collatz Conjecture: Implement a while loop that takes a positive integer n. If n is even, divide it by 2. If n is odd, multiply it by 3 and add 1. Repeat this process until n becomes 1. Print the sequence of numbers.

num = int(input("Enter a number "))
seq = [num]
while num > 1:
    if num % 2 == 0:
        num = num // 2
        seq.append(num)
    else:
        num = num * 3 + 1
        seq.append(num)

print(f"The sequence of Collatz Conjecture is {seq}")
```

```python
# Q11 Pattern Printing
'''
*****
*   *
*   *
*   *
***** 
'''
num = int(input("Enter the number of Rows"))
for i in range(num):
    if i == 0 or i == num - 1:
        print("*" * num)
    else:
        print("*" + (" " * (num - 2)) + "*")
```

```python
# Q12 The Vowel Counter 

num = input("Enter a sentence ")
vowels = set("aeiouAEIOU")
counter = sum(1 for ch in num if ch in vowels)
print(f"The No. of Vowels in '{num}' are {counter}")
```

```python
# Q13 Find the Champion (No Built-ins): Given a list of numbers (e.g., [45, 22, 89, 12, 99, 34]), write a for loop to find the largest number without using Python's built-in max() function.

num = input("Enter a list of numbers separated by spaces")
user_list = list(map(int, num.split()))
largest = user_list[0]
for i in user_list[1:]:
    if i >= largest:
        largest = i
print(f"The Largest number among the given list {user_list} is {largest}")
```

```python
# Q15 Three Strikes Login: Create a while loop that simulates a login screen. Set a predefined password variable. Ask the user to input the password. They get exactly 3 attempts. If they get it right, print "Access Granted" and stop the loop. If they fail 3 times, print "Account Locked".

pswd = "Secret123"
count = 3
while count > 0:
    trial = input("Enter the Password: ")
    if trial == pswd:
        print("Access Granted!")
        break
    else:
        count -= 1
        print(f"Wrong Password entered...{count} attempts left")
        if count == 0:
            print("Account Locked")

```
